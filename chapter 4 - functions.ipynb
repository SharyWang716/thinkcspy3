```python
import turtle
def draw_square (t, sz):
    """make turtle t draw a square of sz"""
    for i in range(4):
        t.forward(sz)
        t.left(90)
        
wn = turtle.Screen()
wn.bgcolor("lightgreen")
wn.title("alex meets a function")

alex = turtle.Turtle()
draw_square(alex, 50)
wn.mainloop()
```

```python
def final_amt(p, r, n, t):
    """apply the compound interest formula to p to produce the final amount"""
    
    a = p*(1+ r/n)**(n*t)
    return a 

toInvest = float(input("how much do you want to invest?"))
fnl = final_amt(toInvest, 0.08, 12, 5)
print("at the end of the period you'll have", fnl)
```

```python
import turtle
def make_window(colr):
    w = turtle.Screen()
    w.bgcolor(colr)
    return w

def make_successive_square(colr, sz):
    t = turtle.Turtle()
    t.color(colr)
    t.pensize(sz)
    return t

wn = make_window("lightgreen")
tess = make_successive_square("hotpink", 3)

size = 20
for i in range(5):
    size = size + 20
    
    for j in range(4):
        tess.forward(size)
        tess.left(90)
        
    tess.penup()
    tess.left(180)
    tess.forward(10)
    tess.left(90)
    tess.forward(10)
    tess.left(90)
    tess.pendown()

wn.mainloop()
```

```python
import turtle
def draw_poly (t, n, sz):
    """make turtle t draw a polygon of sz"""
    for i in range(n):
        t.color("hotpink")
        t.pensize(3)
        t.forward(sz)
        t.left(360/n)
        
        
wn = turtle.Screen()
wn.bgcolor("lightgreen")
#wn.title("alex meets a function")

tess = turtle.Turtle()
draw_poly(tess, 8, 50)
wn.mainloop()
```

```python
import turtle
def draw_pattern (t, n, sz):
    """make turtle t draw a polygon of sz"""
    for i in range(n):
        for j in range(4):
            t.forward(sz)
            t.left(90)
        t.left(360/n)
        
wn = turtle.Screen()
wn.bgcolor("lightgreen")
wn.title("Pretty pattern")

tess = turtle.Turtle()
tess.color("blue")
tess.pensize(3)
draw_pattern(tess, 20, 50)
wn.mainloop()
```

```python
import turtle
def draw_spiral_A (t, n):
    size = 1
    for i in range(n):
        size = size + 2
        t.left(-90)
        t.forward(size)
    
        
def draw_spiral_B(t, n):
    t.penup()
    t.forward(400)
    t.pendown()
    
    size = 1
    for j in range(5):
        size = size + 2
        t.left(-90)
        t.forward(size)
    
    size_s = 9
    for k in range(n):
        size_s = size_s + 2
        t.left(-89)
        t.forward(size_s)
        
        
wn = turtle.Screen()
wn.bgcolor("lightgreen")

tess = turtle.Turtle()
tess.color("blue")
tess.pensize(2)
tess.speed(0)
draw_spiral_A(tess, 103)

alex = turtle.Turtle()
alex.color("blue")
alex.pensize(2)
alex.speed(0)
draw_spiral_B(alex, 103)

wn.mainloop()
```
